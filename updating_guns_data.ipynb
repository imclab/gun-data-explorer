```python
all_cycles = pd.read_sql("SELECT * FROM tmp_bl_gun_groups_spending_all_cycles "
                         "where recipient_state is not null and "
                         "recipient_state != '';",conn)
```

```python
all_cycles.head()
```

```python
all_cycles.transaction_namespace = all_cycles.transaction_namespace.map(
                                    {'urn:fec:transaction':'federal',
                                     'urn:nimsp:transaction':'state'})
```

```python
all_cycles.name = all_cycles.name.map(
                    {'GUN RIGHTS':'PRO-GUN', 
                     'GUN CONTROL':'ANTI-GUN'})
```

```python
all_cycles.head()
```

```python
ac_groups = all_cycles.groupby(by=['transaction_namespace','name','recipient_state']) 
ac_agg = ac_groups.aggregate({'total_amount':np.sum})
ac_agg.head()
```

```python
ac_pivot = ac_agg.unstack('transaction_namespace')
ac_pivot = ac_pivot.unstack('name')
ac_pivot.head()
```

```python
ac_pivot.columns = map(lambda x: x[1]+'_'+x[2], ac_pivot.columns.values)
```

```python
ac_pivot.fillna(value=0.0, inplace=True)
ac_pivot.head()
```

```python
ac_pivot['ANTI-GUN'] = ac_pivot['federal_ANTI-GUN'] + ac_pivot['state_ANTI-GUN']
ac_pivot['PRO-GUN'] = ac_pivot['federal_PRO-GUN'] + ac_pivot['state_PRO-GUN']
```

```python
ac_pivot.head()
```

```python
orig_data.head()
```

```python
state_info = state_info[state_info.fips.str.len() > 0]
state_info.set_index('abbreviation', drop=False, inplace=True)
```

```python
state_info.head()
```

```python
new_state_data = state_info.join(ac_pivot, 
                                 how='left')
```

```python
new_state_data = new_state_data.rename(columns={'abbreviation':'abbr', 
                                                'fips':'id'})
```

```python
new_state_data.fillna(value=0.0, inplace=True)
```

```python
new_state_data.head()
```

```python
new_state_data.reset_index(inplace=True, drop=True)
```

```python
new_state_data.head()
```

```python
new_state_data.to_csv('state_data_update.csv',index=False)
```
